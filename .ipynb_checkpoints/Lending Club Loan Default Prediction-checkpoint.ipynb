# Lending Club Loan Default Prediction

### __Introduction__

Lending Club is platform for connecting borrowers and investors in United Sates, providing access to money for former and opportunity for profit for latter.  

Task at hand involves creating model predicting probability of loan default and through it help investors avoid unnecessary risks as well as allow borrowers to assess their financial viability.  

Additional value can be provided by explaining model's choices - both in terms of validating used features against common sense as well as outlining undesired financial behaviours for applicants.  

Dataset can be found at:  
https://www.kaggle.com/wendykan/lending-club-loan-data/downloads/lending-club-loan-data.zip/1  
Lending Club website with additional information regarding functioning of system:  
https://www.lendingclub.com/

XLS data dictionary file provides description of variables included in dataset (can be opened in other software, purposefully not included in this notebook as not to bloat output). Among them there are those that directly describe actions taken after loan defaulted or was charged off (such as debt settlement status or post charge-off recoveries) so they should be removed.

### __Preprocessing__

Removing singleton columns:

```python
def drop_singleton_columns(df, columns_to_drop=None, drop_type=object):
    '''
    Drops columns from dataframe, can drop predefined ones if columns_to_drop specified. 
    Otherwise checks for columns with only 1 unique value and removes them.
    
    Parameters
    ----------
    df: pandas.DataFrame from which to remove columns
    columns_to_drop: list of columns to drop
    drop_type: dtype of pandas.DataFrame columns to check for singleton columns
    
    Returns
    -------
    df: pandas.DataFrame with dropped columns
    dropped_columns: list of dropped columns
    
    '''
    
    dropped_columns = []
    
    if columns_to_drop:
        df = df.drop(columns=columns_to_drop)
        dropped_columns.append(columns_to_drop)
    
    drop_type_columns = df.loc[:, df.dtypes==drop_type].columns
    
    for col in drop_type_columns:
        
        unique_values = len(list(df[col].unique()))
        if unique_values == 1:
            df = df.drop(columns=col)
            dropped_columns.append(col)
    
    return df, dropped_columns

def cols_with_more_unique_values(df):
    
    object_cols = df.loc[:, df.dtypes==object].columns
    
    more_unique = []
    for col in object_cols:
        if df[col].nunique() > 15:
            more_unique.append(col)
    
    return more_unique
```

Map categorical to numerical:

```python
def map_predicted_class(df):
    
    positive_loans = ['Fully Paid', 'Does not meet the credit policy. Status:Fully Paid']
    negative_loans = ['Charged Off', 'Late (31-120 days)', 'In Grace Period', 'Late (16-30 days)',\
                      'Does not meet the credit policy. Status:Charged Off', 'Default']

    df['loan_classes'] = df.loan_status.apply(lambda x: 1 if x in positive_loans else 0)
    df.drop(columns=['loan_status'], inplace=True)
    
    return df

def map_emp_length(df):
    
    emp_length_map = {'< 1 year':0, '1 year':1, '2 years':2, '3 years':3, '4 years':4, '5 years': 5, '6 years':6\
                      , '7 years':7, '8 years':8, '9 years':9, '10+ years':10}

    df['emp_length'] = df['emp_length'].map(emp_length_map)

    return df
```

Removing columns with more than 65% missing values

```python
def drop_cols_with_missing_threshold(df, thresh_val=df.shape[0]):
    
    cols_before = df.shape[1]
    df = df.dropna(thresh=thresh_val, axis=1)
    print('Removed {} columns '.format(cols_before - df.shape[1]))
    
    return df

#drop_cols_with_missing_threshold(df, thresh_val=int(df.shape[0]*0.65) )
```

#### Imputation of missing values

```python
def simple_imputation(df):
    

    imp = SimpleImputer(strategy='median')
    
    not_obj_cols = df.columns[df.dtypes != object]
    missing_columns = df[not_obj_cols].columns[df[not_obj_cols].isnull().any()]
    imp.fit(df[missing_columns].values)

    df[missing_columns] = imp.transform(df[missing_columns].values)
    
    obj_cols = df.columns[df.dtypes == object]
    object_miss_col = df[obj_cols].columns[df[obj_cols].isnull().any()]
    if object_miss_col:
        mode = df[object_miss_col].mode()
        df[object_miss_col].fillna(mode, inplace=True)
    
    return df
```

Main pre-processing

```python
def process_dataframe(df, train=True):
    
    # dropping columns
    unfair_columns = ['recoveries', 'collection_recovery_fee', 'debt_settlement_flag', 'debt_settlement_flag_date', \
                      'settlement_status', 'settlement_date', 'settlement_amount', 'settlement_percentage', \
                      'settlement_term', 'hardship_flag', 'hardship_type', 'hardship_reason', 'hardship_status', \
                      'deferral_term', 'hardship_amount', 'hardship_start_date', 'hardship_end_date', \
                      'payment_plan_start_date', 'hardship_length', 'hardship_dpd', 'hardship_loan_status', \
                      'orig_projected_additional_accrued_interest', 'hardship_payoff_balance_amount', \
                      'hardship_last_payment_amount', 'last_pymnt_d', 'last_pymnt_amnt', 'last_credit_pull_d', 'pymnt_plan']
    more_to_drop = ['title', 'emp_title', 'issue_d', 'earliest_cr_line', 'mths_since_recent_inq', 'sub_grade', 'zip_code', 'addr_state'] + unfair_columns
    df, _ = drop_singleton_columns(df, columns_to_drop = more_to_drop)
    
    more_uniques = cols_with_more_unique_values(df)
    df.drop(columns=more_uniques, axis=1, inplace=True)
    
    df = drop_cols_with_missing_threshold(df, thresh_val=int(df.shape[0]*0.65) )
    
    # mapping from categorical to numerical
    df = map_predicted_class(df)
    df = map_emp_length(df)
    df['verification_status'] = df.verification_status.map({'Not Verified':0, 'Source Verified':1, 'Verified':1})
    g = {}
    for v in df.grade.unique():
        g[v] = (ord(v) - 65)
    df['grade'] = df.grade.map(g)

    # drop rows for training
    if train==True:
        df = df.query('annual_inc < 500000')
    
    # imputation of missing values
    df = df['emp_length'].fillna(0)
    df = simple_imputation(df)
    
    
    return df
```

```python
df = process_dataframe(df, train=False)
```

#### Encoding

Label encoding:

```python
def label_encoding(df):
    '''
    Function for encoding columns with only 2 unique values by sklearn Label Encoder
    
    Parameters
    ----------
    df : data frame with columns to encode
    
    Returns
    -------
    df : data frame with encoded columns
    encoded_columns: list of encoded columns
    '''
    
    try:
        le = LabelEncoder()
    except:
        from sklearn.preprocessing import LabelEncoder
        le = LabelEncoder()
    
    object_cols = df.loc[:, df.dtypes==object].columns
    encoded_columns = []
    
    for col in object_cols:
        
        unique_values = len(list(df[col].unique()))
        
        if unique_values == 2:
            
            df[col] = le.fit_transform(df[col])
            encoded_columns.append(col)
            
            
    return df, encoded_columns
```

```python
model_log.fit(X_train, y_train)
pred = model_log.predict(X_test)

def metrics(y, pred, train='train'):

    print('ROC/AUC score on {} set is {}'.format(train, roc_auc_score(y, pred)))
    print('F1 score in {} set is {}'.format(train, f1_score(y, pred)))
    
    print('Confusion matrix on {} set:'.format(train))
    print(confusion_matrix(y, pred))
    print('Classification report on {} set:'.format(train))
    print(classification_report(y, pred))

    pass

metrics(y_test, pred, train='test')
```

```python
train_pred = model_log.predict(X_train)
metrics(y_train, train_pred, train='train')
```

```python
model_log_sparse.fit(X_train, y_train)
sparse_pred = model_log_sparse.predict(X_test)

metrics(y_test, sparse_pred, train='test')
```
